```python
df = df.drop(['education', 'fnlwgt'], axis = 1)
df.head()
```

```python
df.replace('?', np.NaN, inplace = True)
df.head()
```

```python
df.fillna(method = 'ffill', inplace = True)
df.reset_index(drop=True, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['workclass'] = le.fit_transform(df['workclass'])
df['marital.status'] = le.fit_transform(df['marital_status'])
df['occupation'] = le.fit_transform(df['occupation'])
df['relationship'] = le.fit_transform(df['relationship'])
df['race'] = le.fit_transform(df['race'])
df['sex'] = le.fit_transform(df['sex'])
df['native_country'] = le.fit_transform(df['native_country'])
df['income'] = le.fit_transform(df['income'])
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Identify categorical features
categorical_features = df.select_dtypes(include=['object']).columns

# One-Hot Encoding for features with <= 5 categories
df_one_hot_encoded = pd.get_dummies(df, columns=[col for col in categorical_features if df[col].nunique() <= 5])

# Label Encoding for features with > 5 categories
label_encoder = LabelEncoder()
for col in categorical_features:
    if df[col].nunique() > 5:
        df[col] = label_encoder.fit_transform(df[col])

print("\nOne-Hot Encoded Data Sample:")
print(df_one_hot_encoded.head())
print("\nLabel Encoded Data Sample:")
print(df.head())
```

```python
df['age_income_ratio'] = df['age'] / (df['hours-per-week'.replace('-','_')])  # Replace '-' with '_' in the column name
df['capital_gain_loss_diff'] = df['capital_gain'] - df['capital_loss']

print("\nData with New Features:")
print(df[['age_income_ratio', 'capital_gain_loss_diff']].head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Identify skewed feature
sns.histplot(df['capital_gain'], kde=True)
plt.title("Original Distribution of Capital Gain")
plt.show()

# Apply log transformation
df['capital_gain_log'] = np.log1p(df['capital_gain'])  # log(1+x) to handle zero values
sns.histplot(df['capital_gain_log'], kde=True)
plt.title("Transformed Distribution of Capital Gain")
plt.show()
```

```python
from sklearn.ensemble import IsolationForest
numerical_features = df.select_dtypes(include=np.number).columns.tolist()

# Detect and remove outliers using Isolation Forest
isolation_forest = IsolationForest(contamination=0.01, random_state=42)
outliers = isolation_forest.fit_predict(df[numerical_features])
df_cleaned = df[outliers == 1]

print(f"\nData shape before outlier removal: {df.shape}")
print(f"Data shape after outlier removal: {df_cleaned.shape}")
```
